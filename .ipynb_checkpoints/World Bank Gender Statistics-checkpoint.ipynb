Questions:
- How does gender equality in terms of education and employment look around the world?
- Is there any pattern among low income and high income countries for gender equality?
- Does increased government spending on education have an impact on female literacy rate?

```python
#Importing required libraries
import numpy as np
import pandas as pd
```

```python
#Reading the first dataset
country = pd.read_csv('Gender_Stats_csv/Gender_StatsCountry.csv')
country.head()
```

```python
country.info()
```

```python
#Dropping the non-required columns
country = country[['Country Code','Short Name','Region','Income Group']]
country.head()
```

```python
#Reading the second dataset
stats = pd.read_csv('Gender_Stats_csv/Gender_StatsData.csv')
stats.head()
```

```python
stats.info()
```

```python
#Converting the horizontal dataset into vertical format
idx = ['Country Name','Country Code','Indicator Name','Indicator Code']

multi_index_df = stats.set_index(idx)
multi_index_df
```

```python
stacked_df = multi_index_df.stack(dropna=False)
stacked_df
```

```python
long_df = stacked_df.reset_index()
long_df.rename({'level_4':'Year', 0:'Indicator Value'}, axis=1, inplace=True)
long_df
```

```python
long_df['Year'].unique()
```

```python
long_df = long_df.loc[long_df['Year']!='Unnamed: 65']
long_df['Year'] = long_df['Year'].astype(int)
```

```python
stats_df = long_df.drop('Indicator Code', axis=1)
stats_df.head()
```

```python
stats_df = stats_df[stats_df['Year']>=1980]
stats_df
```

```python
join_df = pd.merge(stats_df,country,on='Country Code',how='left')
join_df = join_df[['Country Code','Country Name','Region','Income Group','Indicator Name','Year','Indicator Value']]
join_df = join_df[join_df['Region'].notnull()]
join_df
```

```python
#csv_test = join_df[join_df['Indicator Name']=='Expected Years of School']
#csv_test.to_csv(r'C:\Users\Sinchita Siddiquee\Desktop\Sinchita Siddiquee\Personal Projects\World Bank Gender Statistics\csv_test.csv', index=False, header=True)
```

```python
null_check = {}

for i in join_df['Indicator Name'].unique():
    df = join_df[join_df['Indicator Name']==i]
    total_cnt = df['Indicator Value'].size
    null_cnt = df['Indicator Value'].isnull().sum()
    non_null_cnt = total_cnt - null_cnt
    cnt_list = [total_cnt, null_cnt, non_null_cnt]
    null_check[i] = cnt_list
    
null_check
```

```python
#null_df = pd.DataFrame(list(null_check.items()),columns = ['Indicator Name','Count'])
list(null_check.items())
```

```python
df = join_df[join_df['Indicator Name']=='Expected Years of School']
null_cnt = df['Indicator Value'].isnull().sum()
total_cnt = df['Indicator Value'].size
total_cnt
```

```python
null_cnt
```

```python
join_df = join_df[(join_df['Indicator Name']=="Expected years of schooling, female")|
        (join_df['Indicator Name']=="Expected years of schooling, male")|
        (join_df['Indicator Name']=="Female share of employment in senior and middle management (%)")|
        (join_df['Indicator Name']=="GDP per capita (Current US$)")|
        (join_df['Indicator Name']=="Government expenditure on education, total (% of GDP)")|
        (join_df['Indicator Name']=="Gross graduation ratio, primary, female (%)")|
        (join_df['Indicator Name']=="Gross graduation ratio, primary, male (%)")|
        (join_df['Indicator Name']=="Gross graduation ratio, tertiary, female (%)")|
        (join_df['Indicator Name']=="Gross graduation ratio, tertiary, male (%)")|
        (join_df['Indicator Name']=="Labor force, female (% of total labor force)")|
        (join_df['Indicator Name']=="Primary completion rate, female (% of relevant age group)")|
        (join_df['Indicator Name']=="Primary completion rate, male (% of relevant age group)")|
        (join_df['Indicator Name']=="Proportion of seats held by women in national parliaments (%)")|
        (join_df['Indicator Name']=="Self-employed, female (% of female employment) (modeled ILO estimate)")|
        (join_df['Indicator Name']=="Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)")
       ]
```

```python
join_df['Indicator Name'].unique()
```

```python
ind_pivot = join_df.pivot_table('Indicator Value', ['Country Code','Country Name','Region','Income Group','Year'], 'Indicator Name')
ind_pivot.head()
```

```python
ind_pivot.reset_index(inplace=True)
ind_pivot
```

```python
ind_pivot.rename_axis(None, axis=1, inplace=True)
ind_pivot
```

```python
ind_pivot.rename({
'Expected years of schooling, female' : 'Expected Years of Schooling (Female)',
'Expected years of schooling, male' : 'Expected Years of Schooling (Male)',
'Female share of employment in senior and middle management (%)' : 'Share in Senior Management (Female, %)',
'Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)' : 'Share of STEM Graduates (Female, %)',
'GDP per capita (Current US$)' : 'GDP Per Capita (Current US$)',
'Government expenditure on education, total (% of GDP)' : 'Govement Expenditure on Education (% of GDP)',
'Gross graduation ratio, primary, female (%)' : 'Gross Primary Graduation Ratio (Female, %)',
'Gross graduation ratio, primary, male (%)' : 'Gross Primary Graduation Ratio (Male, %)',
'Gross graduation ratio, tertiary, female (%)' : 'Gross Tertiary Graduation Ratio (Female, %)',
'Gross graduation ratio, tertiary, male (%)' : 'Gross Tertiary Graduation Ratio (Male, %)',
'Labor force, female (% of total labor force)' : 'Share of Labor Force (Female, %)',
'Primary completion rate, female (% of relevant age group)' : 'Primary Completion Rate (Female, %)',
'Primary completion rate, male (% of relevant age group)' : 'Primary Completion Rate (Male, %)',
'Proportion of seats held by women in national parliaments (%)' : 'Share of Seats in National Parliament (Female, %)',
'Ratio of female to male labor force participation rate (%) (modeled ILO estimate)' : 'Female To Male Labor Force Participation (%)'
}, axis=1, inplace=True)
```
